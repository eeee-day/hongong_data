```python
#request 패키지

import requests

#get방식은 URL을 requests.get() 함수에 전달하면 끝
url = "http://data4library.kr/api/loanItemSrch?format=json&startDt=2021-04-01&endDt=2021-04-30&age=20&authKey={인증키}"
```

```python
r=requests.get(url)
#get함수가 반환하는 것은 API 호출의 결과를 담고 있는 requests 패키지의 Response클래스.""
```



> Python Requests 패키지


*   GET 방식: requests.get()
*POST 방식: requests.post()
*PUT 방식: requests.put()
*DELETE 방식: requests.delete()



> Response 객체

* response.request : 내가 보낸 request 객체에 접근
* response.status_code : 응답 코드(404 찾을 수 없음. 200 정상)
* response.json() : json문자열을 파이썬 객체로 변환
* response.content : 바이너리 타입으로 데이터 받기
* response.text : UTF-8로 인코딩된 문자열 받기(응답의 원본 텍스트)
* response.encoding : encoding 정보 확인

```python
data = r.json()
print(data)
```

```python
data
```

data 딕셔너리를 books 딕셔너리로 변환하기
딕셔너리.jpg
* docs 키와 doc 키는 각각 하나의 리스트와 하나의 딕셔너리만 가지고 있기 때문에 계층구조를 표현하는 것 이외에는 큰 역할이 없다

```python
#데이터의 딕셔너리 구조 변환

books=[]
for d in data['response']['docs']:
  books.append(d['doc'])

#파이썬 리스트 형식으로도 표현할 수 있다
#books = [d['doc']] for d in data['response']['docs']]
```

```python
books
```

```python
#pandas 데이터프레임으로 변환

import pandas as pd

books_df=pd.DataFrame(books)
books_df
```

```python
#데이터프레임을 활용하기 위해 json으로 변환하기

books_df.to_json('20s_best_book.json')
# to_json() 메서드는 유니코드 16진수로 저장하기 때문에 한글을 그대로 저장하게 하려면
# force_ascii=False로 설정해주기
```
