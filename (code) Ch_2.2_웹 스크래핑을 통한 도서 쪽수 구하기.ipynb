> 전체 도서의 쪽수 구하기 순서

    1. 온라인 서점의 검색 결과 페이지 URL 만들기
    2. requests.get() 함수로 검색 결과 페이지의 HTML을 가져오기
    3. 뷰티플수프로 HTML 파싱
    4. 뷰티풀수프의 find() 메서드로 <a> 태그를 찾아 상세 페이지 URL 추출
    5. requests.get() 함수로 다시 도서 상세 페이지의 HTML 가져오기
    6. 뷰티플수프로 HTML 파싱
    7. 뷰티플수프의 find() 메서드로 ‘품목정보’ <div> 태그 찾기
    8. 뷰티플수프의 find_all() 메서드로 ‘쪽수’가 들어있는 <tr>태그 찾기
    9. 앞에서 찾은 테이블의 행에서 get_text() 메서드로 <td> 태그에 들어있는 ‘쪽수’ 가져오기

> ⭐ 코드를 공부하면서 생긴 문제점.

? idx=7 나미야 잡화점의 기적 책의 페이지가 반환되지 않는다

-> 원인 : yes24 홈페이지 상 해당 도서의 isbn이 변경되었음.

-> 해결방법 : 데이터프레임의 idx 7 isbn13=9791167901484으로 변경해준 뒤 실행하였고, 해당 도서의 페이지 수를 정상적으로 반환하였음.


```python
#검색 결과 페이지 가져오기 : gdown 패키지 사용하기

import gdown

gdown.download('https://bit.ly/3q9SZix','20s_best_books.json', quiet=False)
```

```python
import pandas as pd

books_df = pd.read_json('20s_best_books.json')
books_df.head(10) #첫 다섯개 행 호출
```

```python
#문제 해결 코드

books_df.loc[7, 'isbn13']=9791167901484
books_df.head(10)
```

```python
#데이터 프레임 축소하기 (no~isbn13 까지)
books = books_df[['no','ranking','bookname','authors','publisher','publication_year','isbn13']]
```

```python
# ISBN 정수 값을 받아 쪽수를 반환하는 함수 작성

import requests
from bs4 import BeautifulSoup

def get_page_cnt(isbn):
  #Yes24 도서 검색 페이지 URL
  url = 'https://www.yes24.com/Product/Search?domain=BOOK&query={}'
  #URL에 ISBN을 넣어 HTML 불러오기
  r = requests.get(url.format(isbn))
  #HTML 파싱
  soup = BeautifulSoup(r.text, 'html.parser')

  #검색 결과에서 해당 도서 선택
  prd_info = soup.find('a',attrs={'class':'gd_name'})
  if prd_info == None:
    return ''

  #도서 상세페이지 가져오기
  url = 'http://www.yes24.com'+prd_info['href']
  r=requests.get(url)
  soup = BeautifulSoup(r.text, 'html.parser')

  #품목정보 <div> 선택
  prd_detail = soup.find('div',attrs={'id':'infoset_specific'})

  #테이블에 있는 <tr>태그 가져오기
  prd_tr_list = prd_detail.find_all('tr')

  #쪽수가 들어 있는 <th>를 찾아 <td>에 담긴 값 반환
  for tr in prd_tr_list:
    if tr.find('th').get_text() == '쪽수, 무게, 크기':
      return tr.find('td').get_text().split()[0]
  return ''
```

```python
get_page_cnt(9788972756194)
```

```python
#데이터프레임 행 혹은 열에 함수 적용하기 : apply()메서드
top10_books = books.head(10) #가장 인기있는 10개 행만 가져오기
```

```python
def get_page_cnt2(row):
  isbn = row['isbn13']
  return get_page_cnt(isbn)
```

```python
page_count = top10_books.apply(get_page_cnt2, axis=1)
print(page_count)
#apply() 메서드의 첫 번째 매개변수 : 실행할 함수
# 두 번째 매개변수 : axis (1은 행, 0은 열)
```

```python
# 함수를 더욱 간결하게 만들려면?? : lambda 함수 사용
# -> 함수의 이름 없이 한 줄로 쓰는 함수. 코드가 간결해지며, 빠르게 구현 가능
# 람다 함수를 사용하면 apply() 메서드에 행 또는 열을 전달하는지 의도를 분명하게 나타낼 수 있음

page_count2 = top10_books.apply(lambda row : get_page_cnt(row['isbn13']), axis=1)
#                                     매개변수   리턴값                 적용할 행 또는 열
print(page_count2)
```

```python
#데이터프레임과 시리즈 합치기 : merge()함수
#page_count(시리즈)와 top10_books(데이터프레임) 합치기

#page_count의 이름 설정
page_count.name='page_count'
print(page_count)
```

```python
top10_with_page_count = pd.merge(top10_books, page_count, left_index=True, right_index = True)
# left_index=True, right_index = True -> 인덱스를 기준으로 병합하겠다
top10_with_page_count
```
