```python
#그룹별로 모으기 : groupby()메서드
#하나의 책이 두개의 행으로 중복되어 들어가있고, 대출건수가 다르다면, 하나의 행으로 대출건수를 모으기!

count_df = ns_book[['도서명','저자','ISBN','권','대출건수']]
```

```python
group_df = count_df.groupby(by=['도서명','저자','ISBN','권'],dropna=False) #dropna=Fals Nan이 있는 행 삭제하지 않음
loan_count = group_df.sum() #대출건수는 정수이므로 sum함수를 통해 합치기
```

```python
load_count = count_df.groupby(by=['도서명','저자','ISBN','권'],dropna=False).sum()
loan_count.head()
```

```python
#원본 데이터 업데이트하기

#과정 1 - duplicated()메서드로 중복된 행을 True로 표시한 불리언 배열 생성
dup_rows = ns_book.duplicated(subset=['도서명','저자','ISBN','권']) #중복된 행을 True로 표시

#과정 2 - 1번의 불리언 배열 반전시키기
unique_rows = ~dup_rows

#과정 3 - 2번의 불리언 배열을 사용해 원본 배열에서 고유한 행만 선택
ns_book3 = ns_book[unique_rows].copy() #copy 메서드는 데이터프레임의 복사본 생성함
```

```python
#원본 DF 인덱스 설정하기 : set_index()메서드 -지정한 열을 인덱스로 설정하기
ns_book3.set_index(['도서명','저자','ISBN','권'],inplace=True)
#inplace = true 원본 데이터 프레임 직접변경 / false 새로운 데이터 프레임 반환
ns_book3.head()
```

```python
#업데이트 하기 : update()메서드
ns_book3.update(loan_count)
ns_book3.head()
```

```python
#인덱스 열 해제 : reset_index()메서드

ns_book4 = ns_book3.reset_index()
ns_book4.head()
```

```python
############### 일괄 처리 함수 만들기 ##################
#위의 과정을 일괄로 처리할 수 있는 함수를 만들어보자!

def data_cleaning(filename):
  #param filename : CSV파일이름

  #파일을 DF로 읽기
  ns_df=pd.read_csv(filename, low_memory=False)
  print(ns_df.head())
  #Nan열 삭제
  ns_book = ns_df.dropna(axis=1,how='all')

  #대출 건수를 합치기 위해 필요한 행 추출(count_df)
  count_df = ns_book[['도서명','저자','ISBN','권','대출건수']]

  #도서명, 저자, ISBN, 권 기준으로 groupby
  loan_count = count_df.groupby(by=['도서명','저자','ISBN','권'],dropna=False).sum() #dropna=Fals Nan이 있는 행 삭제하지 않음

  #원본 데이터프레임에서 중복된 행을 제외, 고유한 행만 추출하여 복사
  dup_rows = ns_book.duplicated(subset=['도서명','저자','ISBN','권']) #중복된 행을 True로 표시
  unique_rows = ~dup_rows
  ns_book3 = ns_book[unique_rows].copy() #copy 메서드는 데이터프레임의 복사본 생성함

  #도서명, 저자, ISBN, 권을 인덱스로 설정
  ns_book3.set_index(['도서명','저자','ISBN','권'],inplace=True)

  #loan_count에 대출건수 업데이트
  ns_book3.update(loan_count)

  #인덱스 재설정
  ns_book4 = ns_book3.reset_index()

  #원본 데이터프레임의 열 순서로 변경
  ns_book4 = ns_book4[ns_book.columns]
  return ns_book4
```

```python
new_ns_book4 = data_cleaning('ns_202104.csv')
ns_book4.equals(new_ns_book4) #두 결과가 같은지 비교
```
