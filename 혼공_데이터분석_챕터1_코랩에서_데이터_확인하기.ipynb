```python
import chardet
with open('남산도서관 장서 대출목록 (2021년 04월).csv', mode = 'rb') as f: #binary 읽기모드(인코딩 형식에 상관없이 파일 오픈 가능)로 파일 오픈
  d=f.readline()
print(chardet.detect(d)) #chardet.detect()함수를 사용하여 인코딩 방식 파악하기
```

```python
#csv 파일을 데이터프레임으로 읽기 : read_csv()함수 사용하기
import pandas as pd #pandas패키지를 pd로 import
df = pd.read_csv('남산도서관 장서 대출목록 (2021년 04월).csv', encoding='EUC-KR',low_memory=False)
#low_memory=False -> 데이터프레임의 한 column에 여러 type의 데이터가 섞여있을 경우 나타나는 오류를 출력하지 않게 함
```

```python
df.head() #데이터프레임의 상위 5행 확인
#unnamed:13 -> csv파일의 각 라인 끝에 (,)로 인해 열이 하나 더 있다고 인식하여 생긴 것
```

*   read_csv() : (,)콤마로 열 구분
*   read_table() : (\t)탭으로 열 구분

```python
df.to_csv('ns_202104.csv')#데이터프레임을 csv파일로 저장
```

```python
with open('ns_202104.csv') as f:
  for i in range(3):
    print(f.readline(), end="") #end=""을 이용해 줄바꿈 문자 출력하지 않음
```

```python
ns_df=pd.read_csv('ns_202104.csv',index_col=0,low_memory="false")
ns_df.head()
#unnamed: 0 -> 위에서 생성한 데이터프레임의 index까지 저장하여 열로 인식한 것 .
#따라서, index_col=0을 이용해 0번째 열이 index임을 알려줌
```

*   df.to_csv('ns_202104.csv',index=False) 를 통해 index 매개변수를 저장하지 않는 방법도 있음
*   ns_df.to_excel('ns_202104.xlsx',index=False, engine='xlsxwriter') 를 통해 excel파일로도 저장 가능
